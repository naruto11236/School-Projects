```
train = titanic_df.head(stop)
```

```
stop2 = len(titanic_df)-stop
```

```
print (stop2)
```

```
test = titanic_df.tail(stop2)
```

```
titanic_df.head(stop2)
```

```
from sklearn.neighbors import NearestNeighbors
```

```
X = train
X = X.drop(columns = ["Name","Sex","Ticket","Cabin", "Embarked"])
```

```
X = X[X.Age.isnull() == False]
```

```

nbrs = NearestNeighbors(n_neighbors=3, algorithm='ball_tree').fit(X)
```

```
distances, indices = nbrs.kneighbors(X)
```

```
distances
```

```
indices
```

```

from sklearn.neighbors import KNeighborsClassifier
```

```
clf = KNeighborsClassifier(n_neighbors= 10)
```

```
dir(clf)
```

```
Y = X["Survived"]
```

```
model = clf.fit(X,Y)
```

```
dir(model)
```

```
X_test = test
X_test = X_test.drop(columns = ["Name","Sex","Ticket","Cabin", "Embarked"])
X_test = X_test[X_test.Age.isnull() == False]
```

```
X_test
```

```
Y_test = X_test["Survived"]
```

```
y_pred = model.predict(X_test)
```

```
guess_vector = abs(Y_test - y_pred)
```

```
incorrect_guess = guess_vector.sum()
```

```
acurracy = (len(Y_test)-incorrect_guess)/len(Y_test)
```

```
acurracy
```
